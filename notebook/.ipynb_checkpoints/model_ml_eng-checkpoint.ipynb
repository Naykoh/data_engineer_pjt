```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file ÂµI/O (e.g. pd.read_csv)
import pickle
```

```python
Twitter_df = pd.read_csv('../data/Twitter_Data.csv')
Reddit_df = pd.read_csv('../data/Reddit_Data.csv')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dense,RNN,Bidirectional,Embedding, LSTM, GRU, Flatten, Conv1D, Input,Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.utils import to_categorical
```

```python
# Making the tokenizer ready for using 
tokenizer = Tokenizer()

# Fit the data into the tokenizer: 
tokenizer.fit_on_texts(list(Twitter_df['clean_text']))

text_sequence = tokenizer.texts_to_sequences(list(Twitter_df['clean_text']))
```

```python
max_length = max([len(x) for x in text_sequence])
max_words = len(tokenizer.word_index) + 1
```

```python
pad_sequence = pad_sequences(text_sequence,maxlen = max_length, padding = 'post')
```

```python
labels = to_categorical(Twitter_df['category'],num_classes=3)
```

```python
Reddit_labels = to_categorical(Reddit_df['category'],num_classes=3)
reddit_sequence = tokenizer.texts_to_sequences(list(Twitter_df['clean_text']))
reddit_padded = pad_sequences(reddit_sequence,maxlen = max_length, padding = 'post')
```

```python
def Model():
    model = Sequential()
    model.add(Embedding(max_words, 128, input_length= max_length - 1))
    model.add(Bidirectional(LSTM(128)))
    model.add(Dropout(0.3))
    model.add(Dense(256,activation = 'relu'))
    model.add(Dense(125,activation = 'relu'))
    model.add(Dense(3,activation = 'softmax'))
    return model
```

```python
model = Model()
```

```python
model.compile(optimizer=Adam(),loss=CategoricalCrossentropy(),metrics=['accuracy'])
```

```python
hist = model.fit(reddit_padded,labels,epochs = 1)
```

15h43 lancement

```python
def classifier(text):
    review_seq = tokenizer.texts_to_sequences([text])
    review_padded = pad_sequences(review_seq,maxlen = max_length, padding = 'post')
    prediction = model.predict(review_padded)
    max_classifier = np.argmax(prediction)
    if max_classifier == 0:
        return "neutre"
    elif max_classifier == 1:
        return "positive"
    elif max_classifier == 2:
        return "Negative"
```

```python
model.save('../models')
```

```python
review = "good"
classifier(review)
```

```python
user_tweet = input("Enter your tweet: ")
classing = classifier(user_tweet)
print("The class of your tweet is: ", classing)
```
